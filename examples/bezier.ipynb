```python
import os.path as osp
import sys
from pathlib import Path
import matplotlib.pyplot as plt

sys.path.append(str(Path(osp.abspath("")).parent))

%config InlineBackend.figure_format='retina'
```

```python
from core.curve import Bezier

import numpy as np

control_points = np.array(object=[[110, 150], [25, 190], [210, 250], [210, 30]])

b0 = Bezier(control_points)
p0 = b0.get_points(100)
plt.plot(control_points[:, 0], control_points[:, 1], "ro--", label="Control Points")
plt.plot(p0[:, 0], p0[:, 1], "b-", label="Bezier Curve")
plt.legend()
plt.show()
```

## Bezier

 https://pomax.github.io/bezierinfo/zh-CN/index.html#splitting

```python
plt.plot(control_points[:, 0], control_points[:, 1], "ro--", label="Control Points")
plt.plot(p0[:, 0], p0[:, 1], "r-", label="Bezier Curve")

b1, b2 = b0.split(0.5)

p1 = b1.get_points(20)
left_ctrlpts = b1.control_points
plt.plot(left_ctrlpts[:, 0], left_ctrlpts[:, 1], "bo--", label="Left Control Points")
plt.plot(p1[:, 0], p1[:, 1], "b.", label="Left Curve")

p2 = b2.get_points(20)
right_ctrlpts = b2.control_points
plt.plot(right_ctrlpts[:, 0], right_ctrlpts[:, 1], "go--", label="Right Control Points")
plt.plot(p2[:, 0], p2[:, 1], "g.", label="Right Curve")
plt.legend()
plt.show()
```
